```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\Users\amuke\Desktop\juper\Web_scrapping\phone_under_20K.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Brand']=df['product_name'].str.split(' ',expand=True)[0]
df['Brand_series']=df['product_name'].str.split(' ',expand=True)[1]+df['product_name'].str.split(' ',expand=True)[2]
df['RAM(in GB)']=df['description'].str.split(' ',expand=True)[0]
df['Display_size(in inch)']=df['description'].str.split(' ',expand=True)[8].str.split("(",expand=True)[1]
```

```python
df.head()
```

```python
df['Price'] = df['Price'].replace({'₹': '', ',': '', ' ': ''}, regex=True)
```

```python

df['dispDisplay_size(in inch)lay'] = pd.to_numeric(df['Display_size(in inch)'], errors='coerce')
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

top_5_phones = df.sort_values(by=['Display_size(in inch)', 'Price'], ascending=[False, False]).head(5)

top_5_phones[['Brand', 'Display_size(in inch)', 'Price']]
```

```python
data_cleaned=df.copy()
```

```python
data_cleaned.head()
```

```python
data_cleaned.info()
```

```python
data_cleaned['Price'] = data_cleaned['Price'].astype(int)
```

```python
# 1. Count of unique products
unique_products = data_cleaned['product_name'].nunique()

# 2. Average price of smartphones
average_price = data_cleaned['Price'].mean()

# 3. Minimum and maximum price
min_price = data_cleaned['Price'].min()
max_price = data_cleaned['Price'].max()

# 4. Average rating
average_rating = data_cleaned['Rating'].mean()

# 5. Count of ratings distribution
rating_distribution = data_cleaned['Rating'].value_counts()

# 6. Top 5 most expensive smartphones
top_expensive = data_cleaned.sort_values(by='Price', ascending=False).head(5)

# 7. Top 5 least expensive smartphones
top_cheap = data_cleaned.sort_values(by='Price').head(5)

# 8. Number of products priced above ₹18,000
count_above_18k = data_cleaned[data_cleaned['Price'] > 18000].shape[0]

# 9. Number of products rated 4.4
count_rating_4_4 = data_cleaned[data_cleaned['Rating'] == 4.4].shape[0]

# 10. Products with unique descriptions
unique_description_products = data_cleaned['description'].nunique()

# Presenting insights
insights = {
    "Unique Products": unique_products,
    "Average Price": average_price,
    "Price Range": (min_price, max_price),
    "Average Rating": average_rating,
    "Rating Distribution": rating_distribution.to_dict(),
    "Top 5 Expensive Smartphones": top_expensive[['product_name', 'Price']].to_dict('records'),
    "Top 5 Cheap Smartphones": top_cheap[['product_name', 'Price']].to_dict('records'),
    "Count of Smartphones Above ₹18,000": count_above_18k,
    "Count of Smartphones Rated 4.4": count_rating_4_4,
    "Unique Product Descriptions": unique_description_products
}

insights
```

```python
brand_counts = df['Brand'].value_counts()

# Create the bar plot
plt.figure(figsize=(10,6))
sns.barplot(x=brand_counts.index, y=brand_counts.values, palette='viridis')

# Adding labels and title
plt.xlabel('Phone Brand')
plt.ylabel('Number of Phones')
plt.title('Number of Phones per Brand')

# Show plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# 8. Smartphones Priced Above ₹18,000 (Pie Chart)
plt.figure(figsize=(6, 6))
above_18k = count_above_18k
below_18k = data_cleaned.shape[0] - count_above_18k
labels = ['Above ₹18,000', 'Below ₹18,000']
sizes = [above_18k, below_18k]
colors = ['#FF9999', '#66B2FF']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Smartphones Priced Above ₹18,000', fontsize=14)
plt.show()
```

```python
# 9. Smartphones Rated 4.4 (Pie Chart)
plt.figure(figsize=(6, 6))
rated_4_4 = count_rating_4_4
not_rated_4_4 = data_cleaned.shape[0] - count_rating_4_4
labels = ['Rated 4.4', 'Other Ratings']
sizes = [rated_4_4, not_rated_4_4]
colors = ['#FFD700', '#87CEEB']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Smartphones Rated 4.4', fontsize=14)
plt.show()
```

```python
# 7. Top 5 Least Expensive Smartphones (Bar Chart)
plt.figure(figsize=(10, 5))
sns.barplot(x='Price', y='product_name', data=top_cheap, palette='Blues_d')
plt.title('Top 5 Least Expensive Smartphones', fontsize=14)
plt.xlabel('Price (₹)', fontsize=12)
plt.ylabel('Smartphone', fontsize=12)
plt.show()
```

```python
# 5. Rating Distribution (Count Plot)
plt.figure(figsize=(10, 5))
sns.countplot(x='Rating', data=data_cleaned, palette='viridis')
plt.title('Rating Distribution', fontsize=14)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.show()
```
